```python
import pandas as pd

df1 = pd.read_csv('credits.csv', low_memory=False)
df2 = pd.read_csv('movies_metadata.csv', low_memory=False)

# Convert ID columns to numeric, invalid values become NaN
df1['id'] = pd.to_numeric(df1['id'], errors='coerce')
df2['id'] = pd.to_numeric(df2['id'], errors='coerce')

# Drop rows with NaN id (invalid)
df1 = df1.dropna(subset=['id'])
df2 = df2.dropna(subset=['id'])

# Convert to int for merge
df1['id'] = df1['id'].astype(int)
df2['id'] = df2['id'].astype(int)

df2 = df2.merge(df1, on='id')
print(df2.shape)     # Now row count will be > 0
df2.head()
```

```python
m=df2['vote_count'].quantile(0.9)
m
```

```python
q_movies=df2.copy().loc[df2['vote_count']>=m]
q_movies.shape
```

```python
def weighted_rating(x,m=m,C=C):
    v=x['vote_count']
    R=x['vote_average']
    # Calculation based on the IMDB formula
    return (v/(v+m) * R)+(m/(m+v) * C)
```

```python
# Define a new feature 'score' and calculate its value with weighted_rating()
q_movies['score']=q_movies.apply(weighted_rating,axis=1)
```

```python
# Sort movies based on score calculated above
q_movies=q_movies.sort_values('score',ascending=False)
#Top 15 movies
q_movies[['title','vote_count','vote_average','score']].head(10)
```

```python
from ast import literal_eval

def get_list(x):
    # Convert string representation to python object
    if isinstance(x, str):
        try:
            x = literal_eval(x)
        except:
            return []   # malformed string -> return empty
    
    # If still not list, return empty
    if not isinstance(x, list):
        return []

    result = []
    for item in x:
        # If item is dictionary and contains name
        if isinstance(item, dict) and 'name' in item:
            result.append(item['name'])

    # Return first 3 names if more
    return result[:3]


# APPLY on features
for feature in ['cast', 'genres']:
    df2[feature] = df2[feature].apply(get_list)
```

```python
from ast import literal_eval
import numpy as np

def get_director(x):
    # If the row is a string, convert it into a python object
    if isinstance(x, str):
        try:
            x = literal_eval(x)
        except:
            return np.nan

    # If still not a list, give up
    if not isinstance(x, list):
        return np.nan

    # Otherwise look for director
    for item in x:
        if isinstance(item, dict) and item.get('job') == 'Director':
            return item.get('name')

    return np.nan
```
